## Import libraries and load the graph

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import operator
import powerlaw
```

```python
G = nx.read_edgelist('facebook_combined.txt.gz', create_using=nx.Graph(), nodetype=int)
```

# Basic analysis


```python
print("Number of nodes:", nx.number_of_nodes(G))
print("Number of edges:", nx.number_of_edges(G))
print("Degree average degree:", np.mean([d for _, d in G.degree()]))
print("Diameter of the graph:", nx.diameter(G))
print("Density of the graph:",nx.density(G))
```

```python
fig, ax = plt.subplots(figsize=(20, 13))
ax.axis("off")
plot_options = {"node_size": 5, "with_labels": False, "width": 0.15,"node_color":"green","edge_color":"gray" }
nx.draw_networkx(G, pos=nx.random_layout(G), ax=ax, **plot_options,)
```

# Centrality analysis

## Degree Centrality

```python
dc=nx.degree_centrality(G)
sorted_dc=sorted(dc.items(),key=operator.itemgetter(1),reverse=True)
sorted_dc
```

```python
#plot Degree Centrality Histogram
plt.hist(dc.values())
plt.title("Degree Centrality")
plt.xlabel("Degree Centrality")
plt.ylabel("Counts")
```

## Betweenness Centrality

```python
bc=nx.betweenness_centrality(G)
sorted_bc=sorted(bc.items(),key=operator.itemgetter(1),reverse=True)
sorted_bc
```

```python
plt.hist(bc.values())
plt.title("betweenness Centrality")
plt.xlabel("betweenness Centrality")
plt.ylabel("Counts")
```

## closeness centrality

```python
cc=nx.closeness_centrality(G)
sorted_cc=sorted(cc.items(),key=operator.itemgetter(1),reverse=True)
sorted_cc
```

```python
plt.hist(cc.values())
plt.title("closeness Centrality")
plt.xlabel("closeness Centrality")
plt.ylabel("Counts")
```

# Clustering coefficient analysis

```python
print("average clustering coefficient: ", nx.average_clustering(G))
```

```python
plt.figure(figsize=(20, 12))
plt.hist(nx.clustering(G).values())
plt.title("Clustering Coefficient Histogram ", fontdict={"size": 35})
plt.xlabel("Clustering Coefficient", fontdict={"size": 20})
plt.ylabel("Counts", fontdict={"size": 20})
```

```python
#Count triangles
triangles_per_node = list(nx.triangles(G).values())
print("The number of triangles: ",(sum(triangles_per_node) / 3))
```

# Degree Analysis

```python
# Calculate the degree distribution
degree_sequence = [d for n, d in G.degree()]
degree_hist = nx.degree_histogram(G)

# Plot the degree distribution
plt.plot(degree_hist)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution of the Facebook Graph')
plt.show()
```

```python

# Calculate the mean and standard deviation of the degree distribution
degree_sequence = [d for n, d in G.degree()]
mean_degree = np.mean(degree_sequence)
std_degree = np.std(degree_sequence)

# Print the mean and standard deviation
print("Mean degree:", mean_degree)
print("Standard deviation of degree:", std_degree)

# Calculate the minimum and maximum degrees
min_degree = min(degree_sequence)
max_degree = max(degree_sequence)

# Print the minimum and maximum degrees
print("Minimum degree:", min_degree)
print("Maximum degree:", max_degree)

# Calculate the degree assortativity coefficient
degree_assortativity = nx.degree_assortativity_coefficient(G)

# Print the degree assortativity coefficient
print("Degree assortativity:", degree_assortativity)
```

# Heavy-tail analsis

```python
# Calculate the degree distribution of the graph
degree_sequence = [d for n, d in G.degree()]

# Fit a power-law distribution to the degree distribution data
fit = powerlaw.Fit(degree_sequence)

# Print the results of the power-law fit
print("The power-law exponent is", fit.power_law.alpha)
print("The goodness-of-fit of the power-law distribution is", fit.power_law.sigma)
```

# Comparison with a Watts-Strogatz graph

```python
# Generate a Watts-Strogatz graph with similar parameters
n = G.number_of_nodes()
k = int(2 * G.number_of_edges() / G.number_of_nodes())
p = 0.1
WS = nx.watts_strogatz_graph(n, k, p)

# Calculate the average shortest path length of the Facebook graph
average_shortest_path_length_G = nx.average_shortest_path_length(G)

# Calculate the average shortest path length of the Watts-Strogatz graph
average_shortest_path_length_WS = nx.average_shortest_path_length(WS)

# Calculate the average clustering coefficient of the Facebook graph
average_clustering_G = nx.average_clustering(G)

# Calculate the average clustering coefficient of the Watts-Strogatz graph
average_clustering_WS = nx.average_clustering(WS)

# Print the average shortest path length and average clustering coefficient of the two graphs
print("The average shortest path length of the Facebook graph is", average_shortest_path_length_G)
print("The average shortest path length of the Watts-Strogatz graph is", average_shortest_path_length_WS)
print("The average clustering coefficient of the Facebook graph is", average_clustering_G)
print("The average clustering coefficient of the Watts-Strogatz graph is", average_clustering_WS)
```
